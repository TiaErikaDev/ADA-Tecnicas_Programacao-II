##### **Alterando a base de dados**

```python
#Adicionando a coluna com o valor total dos produtos vendidos
df['Total'] = df['QtdItens'] * df['ValorUnitario']
df.head()
```

```python
#Sparando a data da venda em colunas [Ano][Mes][Dia]
df[['Ano', 'Mes', 'Dia']] = df['DataEmissao'].astype(str).str.split('-', expand = True)
```

```python
#Excluindo colunas desnecessárias
excluir_serie = ['DataEmissao','cdProduto','Produto','cdVendedor']
df.drop(excluir_serie, axis = 1, inplace=True)
```

# **Analisando os dados**

### Faturamento total


*   R$ 17.909.232,00



---





```python
fat = df['Total'].sum().round(0)
```

### Quantidade de vendas


*   52.560

---



```python
qtde_vendas = df['NFe'].value_counts().count()
```

### Ticket médio

*   R$ 340,74

---


```python
ticket = fat / qtde_vendas
ticket
```

### Faturamento por ano

---



```python
df_faturamento = df.groupby('Ano')['Total'].sum().to_frame()
df_faturamento
```

```python
px.bar(df_faturamento, title='Faturamento por ano', height=500, width=800)
```

### Quantidade de vendas por ano

---



```python
qtd_vendas=df.groupby(['Ano']).agg({'NFe':'count'})
```

```python
px.bar(qtd_vendas, title='Quantidade de vendas', height=500, width=800)
```

### Produtos de maior faturamento

---



```python
df_fat_prod = df.groupby('Grupo Produto')['Total'].sum().sort_values(ascending=False).to_frame()
df_fat_prod
```

### Grupo produto vendido por mês

---



```python
df.groupby(['Mes','Grupo Produto']).size().to_frame().head(20)
```

### Relação faturamento por vendedor

---



```python
df_vendedor = df.groupby('Vendedor')['Total'].sum().sort_values(ascending=False).to_frame()
df_vendedor
```

```python
px.bar(df_vendedor, title='Faturamento por vendedor', height=500, width=800)
```

```python
df_linha=df['Linha Produto'].value_counts().to_frame()
px.bar(df_linha, title='Linhas de produtos', height=400, width=500)
```
